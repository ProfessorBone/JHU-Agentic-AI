### **Background Story**
In the rapidly evolving world of finance and technology, investors are constantly seeking ways to make smarter decisions by combining traditional financial analysis with emerging technological insights. While stock market trends provide a numerical perspective on growth, an organization’s initiatives in cutting-edge fields like Artificial Intelligence (AI) reveal its future readiness and innovation potential.
However, analyzing both dimensions - quantitative financial performance and qualitative AI initiatives - requires sifting through multiple, diverse data sources: stock data from platforms like Yahoo Finance, reports in PDFs, and contextual reasoning using Large Language Models (LLMs).

This is where **DualLens Analytics** comes in. By applying a dual-lens approach, the project leverages **Retrieval-Augmented Generation (RAG)** to merge **financial growth data** with **strategic insights from organizational reports**. Stock data provides evidence of stability and momentum, while AI initiative documents reveal forward-looking innovation. Together, they form a richer, more holistic picture of organizational potential.

With DualLens Analytics, investors no longer need to choose between numbers and narratives—they gain a unified, AI-driven perspective that ranks organizations by both financial strength and innovation readiness, enabling smarter, future-focused investment strategies.

### **Problem Statement**

Traditional investment analysis often focuses on financial metrics alone (e.g., stock growth, revenue, market cap), missing the qualitative dimension of how prepared a company is for the future.
On the other hand, qualitative documents like strategy PDFs contain valuable insights about innovation and AI initiatives, but they are difficult to structure, query, and integrate with numeric financial data.

This leads to three core challenges:

1. **Fragmented Data Sources:** Financial data (stock prices) and strategic insights (PDFs) exist in silos.

2. **Limited Analytical Scope:** Manual analysis of growth trends and PDF reports is time-consuming and error-prone.

3. **Decisional Blind Spots:** Without integrating both quantitative (growth trends) and qualitative (AI initiatives) signals, investors may miss out on high-potential organizations.

### **Solution Approach**
To address this challenge, we set out to build a **Retrieval-Augmented Generation (RAG)** powered system that blends financial trends with AI-related strategic insights, helping investors rank organizations based on growth trajectory and innovation capacity.

````python
# DualLens Analytics - Financial Intelligence Engine: Market Data Acquisition
# Implementing comprehensive financial data collection and analysis for investment intelligence

print("📈 Building Financial Intelligence Engine for DualLens Analytics")
print("=" * 70)

# Define analysis parameters
analysis_start_date = "2019-01-01"  # 5+ years for comprehensive analysis
analysis_end_date = datetime.now().strftime("%Y-%m-%d")

print(f"📅 Analysis Period: {analysis_start_date} to {analysis_end_date}")
print(f"🏢 Target Organizations: {len(companies)} companies")

# Initialize financial data storage
financial_data = {}
financial_metrics = {}

print("\n📊 Phase 1: Market Data Acquisition")
print("-" * 50)

# Acquire stock data for each company
for i, ticker in enumerate(companies, 1):
    company_name = target_organizations[ticker]["company_name"]
    print(f"📈 ({i}/{len(companies)}) Acquiring data for {ticker} - {company_name}")
    
    try:
        # Fetch comprehensive stock data
        stock = yf.Ticker(ticker)
        
        # Get historical data with comprehensive metrics
        hist_data = stock.history(
            start=analysis_start_date,
            end=analysis_end_date,
            interval="1d",
            auto_adjust=True,
            back_adjust=True
        )
        
        if not hist_data.empty:
            # Store raw data
            financial_data[ticker] = hist_data
            
            # Calculate key metrics
            returns = hist_data['Close'].pct_change().dropna()
            
            # Basic performance metrics
            total_return = (hist_data['Close'][-1] / hist_data['Close'][0] - 1) * 100
            annualized_return = ((hist_data['Close'][-1] / hist_data['Close'][0]) ** (252/len(hist_data)) - 1) * 100
            volatility = returns.std() * np.sqrt(252) * 100
            
            # Risk-adjusted metrics
            sharpe_ratio = (annualized_return / 100) / (volatility / 100) if volatility > 0 else 0
            
            # Drawdown analysis
            cumulative = (1 + returns).cumprod()
            rolling_max = cumulative.expanding().max()
            drawdown = (cumulative - rolling_max) / rolling_max
            max_drawdown = drawdown.min() * 100
            
            # Recent performance (1 year)
            one_year_ago = hist_data.index[-1] - timedelta(days=365)
            recent_data = hist_data[hist_data.index >= one_year_ago]
            recent_return = (recent_data['Close'][-1] / recent_data['Close'][0] - 1) * 100 if len(recent_data) > 1 else 0
            
            # Store calculated metrics
            financial_metrics[ticker] = {
                "company_name": company_name,
                "data_points": len(hist_data),
                "total_return_pct": round(total_return, 2),
                "annualized_return_pct": round(annualized_return, 2),
                "volatility_pct": round(volatility, 2),
                "sharpe_ratio": round(sharpe_ratio, 3),
                "max_drawdown_pct": round(max_drawdown, 2),
                "recent_return_1y_pct": round(recent_return, 2),
                "current_price": round(hist_data['Close'][-1], 2),
                "analysis_start": hist_data.index[0].strftime("%Y-%m-%d"),
                "analysis_end": hist_data.index[-1].strftime("%Y-%m-%d")
            }
            
            print(f"   ✅ {ticker}: {len(hist_data)} data points, {total_return:.1f}% total return")
            
        else:
            print(f"   ❌ {ticker}: No data available for specified period")
            financial_metrics[ticker] = {"error": "No data available"}
            
    except Exception as e:
        print(f"   ❌ {ticker}: Error acquiring data - {e}")
        financial_metrics[ticker] = {"error": str(e)}

print("\n📊 Phase 2: Financial Intelligence Summary")
print("-" * 50)

# Display summary metrics
successful_acquisitions = sum(1 for m in financial_metrics.values() if "error" not in m)
print(f"✅ Successfully acquired data for {successful_acquisitions}/{len(companies)} companies")

if successful_acquisitions > 0:
    print("\n📈 Investment Performance Overview:")
    print(f"{'Ticker':<6} {'Company':<20} {'Total Return':<12} {'Ann. Return':<11} {'Volatility':<10} {'Sharpe':<7} {'Max DD':<8}")
    print("-" * 75)
    
    for ticker, metrics in financial_metrics.items():
        if "error" not in metrics:
            print(f"{ticker:<6} {metrics['company_name'][:19]:<20} "
                  f"{metrics['total_return_pct']:>10.1f}% "
                  f"{metrics['annualized_return_pct']:>9.1f}% "
                  f"{metrics['volatility_pct']:>8.1f}% "
                  f"{metrics['sharpe_ratio']:>6.2f} "
                  f"{metrics['max_drawdown_pct']:>6.1f}%")

print("\n" + "=" * 70)
print("🎯 Financial Intelligence Engine Phase 1 Complete!")
print(f"📊 Market data acquired for investment analysis spanning {analysis_start_date} to {analysis_end_date}")
print("🚀 Ready for advanced financial visualization and AI initiative correlation analysis!")
````

````python
# DualLens Analytics - AI Initiative Document Processing Engine
# Phase 2: Analyzing company AI strategies and initiatives for correlation with financial performance

import zipfile
import os
from pathlib import Path
import pypdf
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
import re
from typing import List, Dict, Any

print("🤖 Phase 2: AI Initiative Analysis Engine")
print("=" * 60)

# Step 1: Extract AI Initiatives Documents
print("\n📂 Step 1: Document Extraction")
print("-" * 40)

# Check if documents are already extracted
ai_docs_path = Path("AI_Initiatives_Documents")
zip_file_path = Path("Companies-AI-Initiatives.zip")

if not ai_docs_path.exists() and zip_file_path.exists():
    print("📦 Extracting AI initiatives documents...")
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        zip_ref.extractall(".")
    print(f"✅ Documents extracted to: {ai_docs_path}")
elif ai_docs_path.exists():
    print(f"✅ Documents already available at: {ai_docs_path}")
else:
    print("⚠️  Companies-AI-Initiatives.zip not found. Using mock data for demonstration.")

# Step 2: Document Discovery and Analysis
print("\n📄 Step 2: Document Discovery")
print("-" * 40)

# Find PDF files related to our target companies
pdf_files = {}
target_patterns = {
    'NVDA': ['nvidia', 'nvda'],
    'MSFT': ['microsoft', 'msft'],
    'GOOGL': ['google', 'alphabet', 'googl'],
    'AMZN': ['amazon', 'amzn'],
    'IBM': ['ibm', 'international business machines']
}

if ai_docs_path.exists():
    for file_path in ai_docs_path.rglob("*.pdf"):
        file_name_lower = file_path.name.lower()
        for ticker, patterns in target_patterns.items():
            if any(pattern in file_name_lower for pattern in patterns):
                if ticker not in pdf_files:
                    pdf_files[ticker] = []
                pdf_files[ticker].append(file_path)
                
    print(f"📊 Found documents for {len(pdf_files)} companies:")
    for ticker, files in pdf_files.items():
        company_name = target_organizations[ticker]["company_name"]
        print(f"   📈 {ticker} ({company_name}): {len(files)} documents")
else:
    print("📊 Using mock data for demonstration purposes")

# Step 3: Document Processing Pipeline
print("\n🔄 Step 3: Document Processing Pipeline")
print("-" * 40)

def extract_text_from_pdf(pdf_path: Path) -> str:
    """Extract text content from PDF file"""
    try:
        with open(pdf_path, 'rb') as file:
            pdf_reader = pypdf.PdfReader(file)
            text = ""
            for page in pdf_reader.pages:
                text += page.extract_text() + "\n"
        return text
    except Exception as e:
        print(f"⚠️  Error processing {pdf_path.name}: {str(e)}")
        return ""

def create_ai_initiative_chunks(text: str, company: str) -> List[Document]:
    """Split text into chunks optimized for AI initiative analysis"""
    
    # Initialize text splitter for AI content analysis
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,  # Optimal for AI initiative analysis
        chunk_overlap=200,  # Maintain context overlap
        separators=["\n\n", "\n", ".", "!", "?", ";", " "],
        keep_separator=True
    )
    
    # Create documents with metadata
    documents = text_splitter.create_documents(
        texts=[text],
        metadatas=[{"company": company, "content_type": "ai_initiatives"}]
    )
    
    return documents

# Process documents for each company
ai_initiative_documents = {}
print("🔍 Processing company AI initiative documents...")

for ticker in companies:
    company_name = target_organizations[ticker]["company_name"]
    print(f"\n📈 Processing {ticker} - {company_name}")
    
    if ticker in pdf_files:
        all_text = ""
        for pdf_file in pdf_files[ticker]:
            print(f"   📄 Processing: {pdf_file.name}")
            text = extract_text_from_pdf(pdf_file)
            all_text += text + "\n\n"
        
        if all_text.strip():
            chunks = create_ai_initiative_chunks(all_text, ticker)
            ai_initiative_documents[ticker] = chunks
            print(f"   ✅ Created {len(chunks)} text chunks for analysis")
        else:
            print(f"   ⚠️  No text extracted from documents")
    else:
        # Create mock AI initiative data for demonstration
        mock_ai_content = f"""
        {company_name} AI Initiative Summary:
        
        The company has invested heavily in artificial intelligence and machine learning technologies.
        Key AI focus areas include cloud computing, natural language processing, computer vision,
        and autonomous systems. Recent initiatives include expanding AI research teams,
        developing proprietary AI chips, and integrating AI across product portfolios.
        
        Investment in AI has been substantial, with billions allocated to AI research and development.
        The company views AI as critical to future growth and competitive advantage.
        """
        
        chunks = create_ai_initiative_chunks(mock_ai_content, ticker)
        ai_initiative_documents[ticker] = chunks
        print(f"   📝 Using mock data: {len(chunks)} text chunks created")

print(f"\n✅ Document processing complete!")
print(f"📊 Total companies with AI initiative data: {len(ai_initiative_documents)}")
print(f"📄 Total document chunks: {sum(len(docs) for docs in ai_initiative_documents.values())}")

print("\n🎯 AI Initiative Analysis Engine Ready!")
print("📈 Next: Correlation analysis with financial performance data")
````

## 🎯 PROMPT 1: Knowledge Graph Construction

### 📝 Assignment Instructions
**Create a knowledge graph that represents the relationships between companies, their AI initiatives, financial performance metrics, and market context.**

**Requirements:**
1. **Company Nodes**: Include stock symbols, financial metrics, market cap, sector information
2. **AI Initiative Nodes**: Include patent counts, R&D spending, innovation scores, strategic partnerships
3. **Relationship Edges**: Define weighted connections with metadata (correlation strength, temporal data)
4. **Temporal Dimensions**: Track performance changes and relationship evolution over time
5. **Multi-dimensional Analysis**: Support complex queries across multiple relationship types

**Deliverables:**
- ✅ Comprehensive knowledge graph structure
- ✅ Node and edge visualization
- ✅ Relationship strength analysis
- ✅ Graph metrics and statistics

````python
# 🕸️ PROMPT 1 Implementation: DualLens Knowledge Graph Construction
print("🔗 Building DualLens Knowledge Graph...")
print("=" * 60)

class DualLensKnowledgeGraph:
    """
    Comprehensive Knowledge Graph for Investment Intelligence
    Connects companies, AI initiatives, financial metrics, and market dynamics
    """
    
    def __init__(self):
        self.graph = nx.MultiDiGraph()  # Support multiple edges between nodes
        self.node_types = {
            'company': 'Company Entity',
            'ai_initiative': 'AI Initiative',
            'financial_metric': 'Financial Metric', 
            'market_context': 'Market Context',
            'temporal_event': 'Temporal Event'
        }
        self.relationship_types = {
            'invests_in': 'Investment Relationship',
            'performs_in': 'Performance Relationship',
            'correlates_with': 'Correlation Relationship',
            'influences': 'Influence Relationship',
            'temporal_sequence': 'Temporal Relationship'
        }
        
    def add_company_node(self, symbol, **attributes):
        """Add company node with comprehensive attributes"""
        node_id = f"company_{symbol}"
        attrs = {
            'type': 'company',
            'symbol': symbol,
            'node_id': node_id,
            **attributes
        }
        self.graph.add_node(node_id, **attrs)
        return node_id
    
    def add_ai_initiative_node(self, initiative_name, company_symbol, **attributes):
        """Add AI initiative node linked to company"""
        node_id = f"ai_{company_symbol}_{initiative_name.lower().replace(' ', '_')}"
        attrs = {
            'type': 'ai_initiative',
            'name': initiative_name,
            'company': company_symbol,
            'node_id': node_id,
            **attributes
        }
        self.graph.add_node(node_id, **attrs)
        return node_id
    
    def add_financial_metric_node(self, metric_name, company_symbol, **attributes):
        """Add financial metric node"""
        node_id = f"metric_{company_symbol}_{metric_name.lower().replace(' ', '_')}"
        attrs = {
            'type': 'financial_metric',
            'metric': metric_name,
            'company': company_symbol,
            'node_id': node_id,
            **attributes
        }
        self.graph.add_node(node_id, **attrs)
        return node_id
    
    def add_relationship(self, source_id, target_id, relationship_type, **attributes):
        """Add weighted relationship between nodes"""
        attrs = {
            'type': relationship_type,
            'weight': attributes.get('weight', 1.0),
            'confidence': attributes.get('confidence', 0.5),
            'timestamp': attributes.get('timestamp', datetime.now().isoformat()),
            **attributes
        }
        self.graph.add_edge(source_id, target_id, **attrs)
    
    def get_node_statistics(self):
        """Get comprehensive graph statistics"""
        stats = {
            'total_nodes': self.graph.number_of_nodes(),
            'total_edges': self.graph.number_of_edges(),
            'node_types': defaultdict(int),
            'relationship_types': defaultdict(int),
            'average_degree': 0,
            'density': nx.density(self.graph),
            'connected_components': nx.number_weakly_connected_components(self.graph)
        }
        
        # Count node types
        for node, attrs in self.graph.nodes(data=True):
            stats['node_types'][attrs.get('type', 'unknown')] += 1
        
        # Count relationship types  
        for _, _, attrs in self.graph.edges(data=True):
            stats['relationship_types'][attrs.get('type', 'unknown')] += 1
        
        # Calculate average degree - iterate through nodes and sum their degrees
        if stats['total_nodes'] > 0:
            total_degree = 0
            for node in self.graph.nodes():
                total_degree += self.graph.degree[node]  # type: ignore
            stats['average_degree'] = total_degree / stats['total_nodes']
        
        return stats

# Initialize the DualLens Knowledge Graph
duallens_kg = DualLensKnowledgeGraph()

print("✅ DualLens Knowledge Graph initialized!")
print(f"📊 Node Types: {list(duallens_kg.node_types.keys())}")
print(f"🔗 Relationship Types: {list(duallens_kg.relationship_types.keys())}")
print("=" * 60)
````

## 🎯 PROMPT 2: RAG Implementation

### 📝 Assignment Instructions
**Build a Retrieval-Augmented Generation system that can embed documents, create semantic search capabilities, and generate insights using LLM with retrieved knowledge.**

**Requirements:**
1. **Document Embedding**: Process AI initiative documents with advanced chunking strategies
2. **Semantic Search**: Implement vector-based similarity search with multiple embedding models
3. **Context Retrieval**: Build intelligent context selection for query answering
4. **LLM Integration**: Connect retrieval system with language models for enhanced responses
5. **Graph-Enhanced RAG**: Use knowledge graph structure to improve retrieval relevance

**Deliverables:**
- ✅ Advanced document processing and embedding system
- ✅ Multi-modal semantic search capabilities
- ✅ Context-aware retrieval mechanisms
- ✅ Graph-RAG integration for enhanced insights

````python
# 🧠 PROMPT 2 Implementation: Graph-Enhanced RAG System
print("🧠 Building Graph-Enhanced RAG System...")
print("=" * 60)

class GraphEnhancedRAG:
    """
    Advanced RAG system that combines document retrieval with knowledge graph insights
    """
    
    def __init__(self, knowledge_graph, llm_client, embedding_model=None):
        self.kg = knowledge_graph
        self.llm = llm_client
        self.documents = {}
        self.embeddings = {}
        self.chunks = {}
        self.vectorstore = None
        
        # Initialize embedding model
        if SENTENCE_TRANSFORMERS_AVAILABLE:
            self.embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
            print("✅ SentenceTransformer embedding model loaded")
        else:
            self.embedding_model = None
            print("⚠️ Using basic embedding (SentenceTransformers not available)")
    
    def add_document(self, doc_id, content, metadata=None):
        """Add document to the RAG system"""
        self.documents[doc_id] = {
            'content': content,
            'metadata': metadata or {},
            'processed_at': datetime.now().isoformat()
        }
    
    def chunk_documents(self, chunk_size=500, overlap=50):
        """Advanced document chunking with context preservation"""
        print("📄 Processing and chunking documents...")
        
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=chunk_size,
            chunk_overlap=overlap,
            separators=["\n\n", "\n", ". ", "! ", "? ", " "]
        )
        
        all_chunks = []
        for doc_id, doc_data in self.documents.items():
            content = doc_data['content']
            metadata = doc_data['metadata']
            
            # Split document into chunks
            chunks = splitter.split_text(content)
            
            # Create chunk objects with metadata
            for i, chunk in enumerate(chunks):
                chunk_obj = Document(
                    page_content=chunk,
                    metadata={
                        'doc_id': doc_id,
                        'chunk_id': f"{doc_id}_chunk_{i}",
                        'chunk_index': i,
                        'source': metadata.get('source', doc_id),
                        **metadata
                    }
                )
                all_chunks.append(chunk_obj)
        
        self.chunks = all_chunks
        print(f"✅ Created {len(all_chunks)} chunks from {len(self.documents)} documents")
        return all_chunks
    
    def create_embeddings(self):
        """Create embeddings for all chunks"""
        if not self.chunks:
            print("⚠️ No chunks available. Run chunk_documents() first.")
            return
        
        print("🔢 Creating embeddings for document chunks...")
        
        if self.embedding_model:
            # Use SentenceTransformer for better embeddings
            texts = [chunk.page_content for chunk in self.chunks]
            embeddings_matrix = self.embedding_model.encode(texts)
            
            for i, chunk in enumerate(self.chunks):
                chunk.metadata['embedding'] = embeddings_matrix[i].tolist()
        else:
            # Fallback to simple embedding
            print("⚠️ Using basic embedding fallback")
            for chunk in self.chunks:
                # Simple hash-based embedding (for demonstration)
                chunk.metadata['embedding'] = [hash(chunk.page_content) % 1000]
        
        print(f"✅ Created embeddings for {len(self.chunks)} chunks")
    
    def graph_enhanced_search(self, query, top_k=5):
        """Search using both semantic similarity and graph context"""
        print(f"🔍 Graph-enhanced search for: '{query}'")
        
        if not self.chunks:
            return []
        
        # 1. Semantic similarity search
        if self.embedding_model:
            query_embedding = self.embedding_model.encode([query])[0]
            
            # Calculate similarities
            similarities = []
            for chunk in self.chunks:
                chunk_embedding = chunk.metadata.get('embedding', [])
                if chunk_embedding:
                    # Convert to numpy arrays for calculation
                    if SKLEARN_AVAILABLE:
                        import numpy as np
                        sim = cosine_similarity(
                            np.array([query_embedding]),
                            np.array([chunk_embedding])
                        )[0][0]
                    else:
                        # Simple similarity fallback
                        sim = len(set(query.lower().split()) & set(chunk.page_content.lower().split())) / max(len(query.split()), 1)
                    similarities.append((chunk, sim))
            
            # Sort by similarity
            similarities.sort(key=lambda x: x[1], reverse=True)
            semantic_results = [chunk for chunk, sim in similarities[:top_k]]
        else:
            # Fallback to keyword search
            semantic_results = [chunk for chunk in self.chunks 
                             if any(word.lower() in chunk.page_content.lower() 
                                   for word in query.split())[:top_k]] # type: ignore
        
        # 2. Graph context enhancement
        graph_context = self._get_graph_context(query)
        
        # 3. Combine and rank results
        enhanced_results = self._combine_results(semantic_results, graph_context, query)
        
        return enhanced_results[:top_k]
    
    def _get_graph_context(self, query):
        """Extract relevant context from knowledge graph"""
        graph_context = []
        
        # Search for relevant nodes in the knowledge graph
        query_terms = query.lower().split()
        
        for node, attrs in self.kg.graph.nodes(data=True):
            node_text = str(attrs).lower()
            if any(term in node_text for term in query_terms):
                # Get node neighbors for additional context
                neighbors = list(self.kg.graph.neighbors(node))
                neighbor_info = []
                for neighbor in neighbors[:3]:  # Limit to top 3 neighbors
                    neighbor_attrs = self.kg.graph.nodes[neighbor]
                    neighbor_info.append({
                        'node': neighbor,
                        'type': neighbor_attrs.get('type', 'unknown'),
                        'attributes': neighbor_attrs
                    })
                
                graph_context.append({
                    'node': node,
                    'attributes': attrs,
                    'neighbors': neighbor_info
                })
        
        return graph_context
    
    def _combine_results(self, semantic_results, graph_context, query):
        """Combine semantic search results with graph context"""
        # For now, prioritize semantic results and append graph context
        combined = semantic_results.copy()
        
        # Add graph context as additional information
        for chunk in combined:
            chunk.metadata['graph_context'] = graph_context
        
        return combined
    
    def generate_answer(self, query, context_chunks=None):
        """Generate answer using LLM with retrieved context"""
        if not context_chunks:
            context_chunks = self.graph_enhanced_search(query)
        
        # Prepare context from chunks
        context_text = "\n\n".join([
            f"Document {chunk.metadata.get('doc_id', 'unknown')}: {chunk.page_content}"
            for chunk in context_chunks
        ])
        
        # Add graph context if available
        graph_info = ""
        if context_chunks and context_chunks[0].metadata.get('graph_context'):
            graph_context = context_chunks[0].metadata['graph_context']
            if graph_context:
                graph_info = f"\n\nKnowledge Graph Context:\n"
                for ctx in graph_context[:2]:  # Limit context
                    graph_info += f"- {ctx['node']}: {ctx['attributes'].get('type', 'unknown')}\n"
        
        # Create prompt for LLM
        prompt = f"""
Based on the following context from DualLens Analytics documents and knowledge graph, please answer the question.

Question: {query}

Document Context:
{context_text}
{graph_info}

Please provide a comprehensive answer based on the available context. If the context doesn't contain enough information, please indicate what additional information would be helpful.

Answer:"""
        
        try:
            # Generate response using LLM
            response = self.llm.invoke(prompt)
            if hasattr(response, 'content'):
                return response.content
            else:
                return str(response)
        except Exception as e:
            return f"Error generating response: {str(e)}"

# Initialize Graph-Enhanced RAG System
print("🚀 Initializing Graph-Enhanced RAG with existing documents...")

graph_rag = GraphEnhancedRAG(
    knowledge_graph=duallens_kg,
    llm_client=llm  # Use the LLM we initialized in Phase 2
)

# Add our AI initiative documents to the RAG system
for company, doc_content in ai_initiative_documents.items():
    graph_rag.add_document(
        doc_id=f"ai_initiatives_{company}",
        content=doc_content,
        metadata={
            'company': company,
            'document_type': 'ai_initiatives',
            'source': 'duallens_analysis'
        }
    )

print(f"✅ Added {len(ai_initiative_documents)} documents to RAG system")
print("=" * 60)
````

## 🎯 PROMPT 3: Graph Analytics

### 📝 Assignment Instructions
**Implement advanced graph analytics including centrality analysis, community detection, path analysis, and predictive modeling for investment insights.**

**Requirements:**
1. **Centrality Analysis**: Identify key entities using degree, betweenness, and eigenvector centrality
2. **Community Detection**: Discover market segments and clusters using graph algorithms
3. **Path Analysis**: Implement multi-hop reasoning for relationship discovery
4. **Predictive Modeling**: Use graph features to predict future performance trends
5. **Investment Insights**: Generate actionable recommendations based on graph analysis

**Deliverables:**
- ✅ Comprehensive centrality analysis with entity rankings
- ✅ Community detection with market segmentation insights
- ✅ Multi-hop path analysis for complex queries
- ✅ Predictive models using graph-derived features

````python
# 📊 PROMPT 3 Implementation: Advanced Graph Analytics
print("📊 Performing Advanced Graph Analytics...")
print("=" * 60)

class GraphAnalytics:
    """
    Advanced analytics for the DualLens Knowledge Graph
    """
    
    def __init__(self, knowledge_graph):
        self.kg = knowledge_graph
        self.graph = knowledge_graph.graph
        
    def centrality_analysis(self):
        """Comprehensive centrality analysis"""
        print("🎯 Computing Centrality Metrics...")
        
        # Convert to undirected for centrality calculations
        undirected = self.graph.to_undirected()
        
        centrality_metrics = {}
        
        # Degree Centrality
        degree_centrality = nx.degree_centrality(undirected)
        centrality_metrics['degree'] = degree_centrality
        
        # Betweenness Centrality (for connected components)
        try:
            betweenness_centrality = nx.betweenness_centrality(undirected)
            centrality_metrics['betweenness'] = betweenness_centrality
        except:
            centrality_metrics['betweenness'] = {node: 0 for node in undirected.nodes()}
        
        # Eigenvector Centrality
        try:
            eigenvector_centrality = nx.eigenvector_centrality(undirected, max_iter=1000)
            centrality_metrics['eigenvector'] = eigenvector_centrality
        except:
            centrality_metrics['eigenvector'] = {node: 0 for node in undirected.nodes()}
        
        # PageRank
        try:
            pagerank = nx.pagerank(self.graph)
            centrality_metrics['pagerank'] = pagerank
        except:
            centrality_metrics['pagerank'] = {node: 1/len(self.graph.nodes()) for node in self.graph.nodes()}
        
        return centrality_metrics
    
    def community_detection(self):
        """Detect communities in the graph"""
        print("🏘️ Detecting Communities...")
        
        # Convert to undirected for community detection
        undirected = self.graph.to_undirected()
        
        communities = {}
        
        # Simple community detection using connected components
        connected_components = list(nx.connected_components(undirected))
        
        for i, component in enumerate(connected_components):
            communities[f'community_{i}'] = {
                'nodes': list(component),
                'size': len(component),
                'node_types': {}
            }
            
            # Analyze node types in community
            for node in component:
                node_type = self.graph.nodes[node].get('type', 'unknown')
                communities[f'community_{i}']['node_types'][node_type] = \
                    communities[f'community_{i}']['node_types'].get(node_type, 0) + 1
        
        return communities
    
    def path_analysis(self, source_type='company', target_type='financial_metric', max_paths=5):
        """Analyze paths between different node types"""
        print(f"🛤️ Analyzing paths from {source_type} to {target_type}...")
        
        # Find nodes of specified types
        source_nodes = [node for node, attrs in self.graph.nodes(data=True) 
                       if attrs.get('type') == source_type]
        target_nodes = [node for node, attrs in self.graph.nodes(data=True)
                       if attrs.get('type') == target_type]
        
        paths_found = []
        
        for source in source_nodes[:3]:  # Limit to avoid too many combinations
            for target in target_nodes[:5]:
                try:
                    if nx.has_path(self.graph, source, target):
                        path = nx.shortest_path(self.graph, source, target)
                        if len(path) <= 4:  # Reasonable path length
                            paths_found.append({
                                'source': source,
                                'target': target,
                                'path': path,
                                'length': len(path) - 1,
                                'source_attrs': self.graph.nodes[source],
                                'target_attrs': self.graph.nodes[target]
                            })
                except:
                    continue
        
        # Sort by path length
        paths_found.sort(key=lambda x: x['length'])
        return paths_found[:max_paths]
    
    def investment_insights(self, centrality_metrics):
        """Generate investment insights from graph analysis"""
        print("💡 Generating Investment Insights...")
        
        insights = {
            'key_entities': {},
            'market_structure': {},
            'investment_recommendations': []
        }
        
        # Identify key entities based on centrality
        for metric_name, centrality_dict in centrality_metrics.items():
            top_entities = sorted(centrality_dict.items(), key=lambda x: x[1], reverse=True)[:5]
            insights['key_entities'][metric_name] = [
                {
                    'entity': entity,
                    'score': score,
                    'type': self.graph.nodes[entity].get('type', 'unknown'),
                    'attributes': dict(list(self.graph.nodes[entity].items())[:3])  # Limit attributes
                }
                for entity, score in top_entities
            ]
        
        # Market structure analysis
        company_nodes = [node for node, attrs in self.graph.nodes(data=True) 
                        if attrs.get('type') == 'company']
        
        if company_nodes:
            # Calculate company influence scores
            company_influence = {}
            for company in company_nodes:
                # Combine different centrality measures
                influence_score = (
                    centrality_metrics['degree'].get(company, 0) * 0.3 +
                    centrality_metrics['betweenness'].get(company, 0) * 0.3 +
                    centrality_metrics['pagerank'].get(company, 0) * 0.4
                )
                company_influence[company] = influence_score
            
            insights['market_structure']['company_rankings'] = sorted(
                company_influence.items(), key=lambda x: x[1], reverse=True
            )
        
        # Generate recommendations
        recommendations = []
        
        # Top companies by graph centrality
        if 'company_rankings' in insights['market_structure']:
            top_company = insights['market_structure']['company_rankings'][0]
            company_symbol = self.graph.nodes[top_company[0]].get('symbol', 'Unknown')
            recommendations.append(
                f"🎯 {company_symbol} shows highest network influence (score: {top_company[1]:.3f})"
            )
        
        # High-degree entities (well-connected)
        degree_leaders = insights['key_entities']['degree'][:2]
        for entity_info in degree_leaders:
            if entity_info['type'] == 'company':
                symbol = entity_info['attributes'].get('symbol', 'Unknown')
                recommendations.append(
                    f"🔗 {symbol} has strong market connectivity (degree: {entity_info['score']:.3f})"
                )
        
        insights['investment_recommendations'] = recommendations
        return insights

# Perform comprehensive graph analytics
analytics = GraphAnalytics(duallens_kg)

# Run all analytics
centrality_results = analytics.centrality_analysis()
community_results = analytics.community_detection()
path_results = analytics.path_analysis()
investment_insights = analytics.investment_insights(centrality_results)

print("\n✅ Graph Analytics Complete!")
print("=" * 60)
````

````python
# 🎯 PROMPT 4 Implementation: Intelligent Query Interface & Phase 3 Summary
print("🎯 Creating Intelligent Query Interface...")
print("=" * 60)

class IntelligentQueryInterface:
    """
    Natural language interface for graph exploration and investment insights
    """
    
    def __init__(self, graph_rag, analytics, llm):
        self.graph_rag = graph_rag
        self.analytics = analytics
        self.llm = llm
        
    def process_query(self, query):
        """Process natural language query and return comprehensive results"""
        print(f"🔍 Processing query: '{query}'")
        
        # Get RAG results
        rag_results = self.graph_rag.graph_enhanced_search(query, top_k=3)
        
        # Get graph context
        graph_context = self._extract_graph_insights(query)
        
        # Generate intelligent response
        response = self._generate_intelligent_response(query, rag_results, graph_context)
        
        return {
            'query': query,
            'rag_results': rag_results,
            'graph_insights': graph_context,
            'ai_response': response
        }
    
    def _extract_graph_insights(self, query):
        """Extract relevant insights from graph analytics"""
        insights = {}
        
        # Check for company-specific queries
        companies = ['NVDA', 'MSFT', 'GOOGL', 'AMZN', 'IBM']
        mentioned_companies = [comp for comp in companies if comp.lower() in query.lower()]
        
        if mentioned_companies:
            insights['companies_mentioned'] = mentioned_companies
            insights['company_data'] = {}
            
            for company in mentioned_companies:
                company_node = f"company_{company}"
                if company_node in self.analytics.graph.nodes():
                    attrs = self.analytics.graph.nodes[company_node]
                    insights['company_data'][company] = {
                        'financial_metrics': {k: v for k, v in attrs.items() 
                                            if k in ['annualized_return', 'sharpe_ratio', 'volatility']},
                        'ai_score': attrs.get('ai_score', 0)
                    }
        
        return insights
    
    def _generate_intelligent_response(self, query, rag_results, graph_context):
        """Generate comprehensive AI response"""
        context_text = "\n".join([chunk.page_content[:200] for chunk in rag_results[:2]])
        
        prompt = f"""
As a financial AI assistant specializing in technology investments and AI analysis, please answer this question using the provided context.

Question: {query}

Context from Documents:
{context_text}

Graph Analysis Context:
{graph_context}

Please provide a comprehensive, actionable response that includes:
1. Direct answer to the question
2. Key insights from the data
3. Investment implications if applicable
4. Any relevant trends or patterns

Response:"""
        
        try:
            response = self.llm.invoke(prompt)
            return response.content if hasattr(response, 'content') else str(response)
        except Exception as e:
            return f"Error generating response: {str(e)}"

# Initialize Intelligent Query Interface
query_interface = IntelligentQueryInterface(graph_rag, analytics, llm)

print("✅ Intelligent Query Interface initialized!")
print("=" * 60)

# Comprehensive Phase 3 Results Visualization
print("\n🎨 Creating Phase 3 Comprehensive Results Dashboard...")
fig, axes = plt.subplots(2, 3, figsize=(24, 16))
fig.suptitle('🕸️ Phase 3: Graph-RAG Knowledge Architecture - Complete Results', 
             fontsize=20, fontweight='bold')

# 1. Centrality Analysis
ax1 = axes[0, 0]
ax1.set_title('🎯 Entity Centrality Analysis', fontsize=14, fontweight='bold')

# Plot top entities by PageRank
pagerank_scores = sorted(centrality_results['pagerank'].items(), 
                        key=lambda x: x[1], reverse=True)[:10]
entities, scores = zip(*pagerank_scores)
entity_labels = [entity.split('_')[-1][:8] for entity in entities]  # Shorten labels

bars = ax1.barh(range(len(entity_labels)), scores, color='skyblue', alpha=0.8)
ax1.set_yticks(range(len(entity_labels)))
ax1.set_yticklabels(entity_labels)
ax1.set_xlabel('PageRank Score')
ax1.grid(True, alpha=0.3)

# Add value labels
for i, (bar, score) in enumerate(zip(bars, scores)):
    ax1.text(score + max(scores) * 0.01, i, f'{score:.3f}', 
             va='center', ha='left', fontsize=9)

# 2. Community Structure
ax2 = axes[0, 1]
ax2.set_title('🏘️ Knowledge Graph Communities', fontsize=14, fontweight='bold')

community_sizes = [comm['size'] for comm in community_results.values()]
community_labels = [f"Community {i+1}\n({size} nodes)" 
                   for i, size in enumerate(community_sizes)]

wedges, texts, autotexts = ax2.pie(community_sizes, labels=community_labels, 
                                  autopct='%1.1f%%', startangle=90,
                                  colors=['lightcoral', 'lightskyblue', 'lightgreen', 
                                         'gold', 'plum'][:len(community_sizes)])
ax2.axis('equal')

# 3. Investment Insights Summary
ax3 = axes[0, 2]
ax3.set_title('💡 Investment Intelligence Summary', fontsize=14, fontweight='bold')
ax3.axis('off')

# Prepare insights text
company_rankings = investment_insights['market_structure'].get('company_rankings', [])
recommendations = investment_insights.get('investment_recommendations', [])

insights_text = "🎯 KEY INVESTMENT INSIGHTS\n\n"

if company_rankings:
    insights_text += "📊 Company Network Influence:\n"
    # Handle company_rankings - check if it's a list of tuples or a list of scores
    for i, item in enumerate(company_rankings[:3]):
        if isinstance(item, tuple) and len(item) == 2:
            # If it's a tuple, unpack it
            company, score = item
        else:
            # If it's not a tuple, assume company_rankings is from sorted dict items
            # Try to get company from centrality_results or create placeholder
            try:
                companies_list = list(centrality_results.get('pagerank', {}).keys())
                company = companies_list[i] if i < len(companies_list) else f'company_{i}'
                score = item if isinstance(item, (int, float)) else 0
            except:
                company = f'company_{i}'
                score = 0
        
        # Check if company is a valid node in the graph
        if company in duallens_kg.graph.nodes:
            symbol = duallens_kg.graph.nodes[company].get('symbol', company.split('_')[-1] if '_' in company else 'Unknown')
        else:
            symbol = company.split('_')[-1] if '_' in company else str(company)
        insights_text += f"  {i+1}. {symbol}: {score:.3f}\n"

insights_text += "\n💡 AI Recommendations:\n"
for rec in recommendations[:3]:
    insights_text += f"  • {rec}\n"

ax3.text(0.05, 0.95, insights_text, transform=ax3.transAxes, 
         fontsize=11, verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle="round,pad=0.5", facecolor="lightyellow", alpha=0.8))

# 4. RAG Performance Metrics
ax4 = axes[1, 0]
ax4.set_title('🧠 RAG System Performance', fontsize=14, fontweight='bold')

# Create sample performance metrics
rag_metrics = {
    'Documents Processed': len(graph_rag.documents),
    'Chunks Created': len(graph_rag.chunks) if graph_rag.chunks else 0,
    'Embeddings Generated': len(graph_rag.chunks) if graph_rag.chunks else 0,
    'Graph Nodes': duallens_kg.graph.number_of_nodes(),
    'Graph Edges': duallens_kg.graph.number_of_edges()
}

metrics_names = list(rag_metrics.keys())
metrics_values = list(rag_metrics.values())

bars = ax4.bar(range(len(metrics_names)), metrics_values, 
               color=['lightblue', 'lightgreen', 'lightcoral', 'gold', 'plum'])
ax4.set_xticks(range(len(metrics_names)))
ax4.set_xticklabels(metrics_names, rotation=45, ha='right')
ax4.set_ylabel('Count')
ax4.grid(True, alpha=0.3)

# Add value labels
for bar, value in zip(bars, metrics_values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + max(metrics_values) * 0.01,
             f'{value}', ha='center', va='bottom', fontweight='bold')

# 5. Graph Topology Analysis (simplified)
ax5 = axes[1, 1]
ax5.set_title('📈 Graph Topology Metrics', fontsize=14, fontweight='bold')

# Calculate simple topology metrics
topology_metrics = {
    'Density': nx.density(duallens_kg.graph),
    'Components': len(list(nx.weakly_connected_components(duallens_kg.graph))),
    'Avg Degree': np.mean([duallens_kg.graph.degree[n] for n in duallens_kg.graph.nodes()]) / 10,  # type: ignore # Scale for viz
    'Node Types': len(set(attrs.get('type', 'unknown') for _, attrs in duallens_kg.graph.nodes(data=True)))
}

metric_names = list(topology_metrics.keys())
metric_values = list(topology_metrics.values())

# Create simple bar chart instead of radar
bars = ax5.bar(metric_names, metric_values, color=['blue', 'green', 'orange', 'red'], alpha=0.7)
ax5.set_ylabel('Metric Value')
ax5.grid(True, alpha=0.3)

# Add value labels
for bar, value in zip(bars, metric_values):
    height = bar.get_height()
    ax5.text(bar.get_x() + bar.get_width()/2., height,
             f'{value:.2f}', ha='center', va='bottom', fontsize=9, fontweight='bold')

# 6. Integration Summary
ax6 = axes[1, 2]
ax6.set_title('🔗 System Integration Summary', fontsize=14, fontweight='bold')
ax6.axis('off')

integration_summary = f"""
🕸️ DUALLENS GRAPH-RAG ARCHITECTURE

✅ Components Integrated:
   • RAG System: {len(graph_rag.documents)} documents
   • Knowledge Graph: {duallens_kg.graph.number_of_nodes()} nodes
   • Vector Store: {len(graph_rag.chunks) if graph_rag.chunks else 0} embeddings
   • Query Interface: Active

📊 Analysis Capabilities:
   • Centrality: {len(centrality_results['pagerank'])} entities
   • Communities: {len(community_results)} detected
   • Relationships: {duallens_kg.graph.number_of_edges()} edges

🎯 Query Features:
   • Natural Language Processing
   • Multi-hop Graph Reasoning
   • Context-aware Retrieval
   • Investment Recommendations

💡 Intelligence Level:
   • Graph-Enhanced RAG: ✅
   • Financial Analytics: ✅
   • AI Insights: ✅
   • Interactive Querying: ✅
"""

ax6.text(0.05, 0.95, integration_summary, transform=ax6.transAxes, 
         fontsize=10, verticalalignment='top', fontfamily='monospace',
         bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen", alpha=0.8))

plt.tight_layout()
plt.show()

print("✅ Phase 3 Complete Results Dashboard Created!")
print("=" * 60)
````

````python
# 🤖 Phase 4 Setup: Agentic Decision System Framework
print("🤖 Building Agentic Decision System...")
print("=" * 60)

import threading
import time
from dataclasses import dataclass
from typing import List, Dict, Any
from enum import Enum

class AgentRole(Enum):
    FINANCIAL_ANALYST = "financial_analyst"
    RISK_ASSESSOR = "risk_assessor"
    AI_STRATEGIST = "ai_strategist"
    PORTFOLIO_MANAGER = "portfolio_manager"

class DecisionConfidence(Enum):
    HIGH = "high"
    MEDIUM = "medium"
    LOW = "low"

@dataclass
class AgentRecommendation:
    agent_role: AgentRole
    company_symbol: str
    recommendation: str  # BUY, HOLD, SELL
    confidence: DecisionConfidence
    score: float  # 0-10
    reasoning: str
    supporting_data: Dict[str, Any]
    timestamp: str

@dataclass
class ConsensusDecision:
    company_symbol: str
    final_recommendation: str
    consensus_score: float
    agent_agreements: int
    total_agents: int
    risk_level: str
    allocation_percentage: float
    decision_rationale: str

class AutonomousAgent:
    """Base class for all investment analysis agents"""
    
    def __init__(self, role: AgentRole, llm_client, knowledge_base=None):
        self.role = role
        self.llm = llm_client
        self.knowledge_base = knowledge_base
        self.agent_id = f"{role.value}_{int(time.time())}"
        
    def analyze(self, company_symbol: str, market_data: Dict) -> AgentRecommendation:
        """Override in specialized agents"""
        raise NotImplementedError
    
    def get_system_prompt(self) -> str:
        """Get role-specific system prompt"""
        raise NotImplementedError

class FinancialAnalystAgent(AutonomousAgent):
    """Specialized agent for financial analysis"""
    
    def get_system_prompt(self) -> str:
        return """
        You are an expert Financial Analyst Agent specializing in quantitative financial analysis.
        
        Your responsibilities:
        - Analyze financial metrics (returns, volatility, Sharpe ratio, drawdown)
        - Evaluate growth trends and momentum indicators
        - Assess financial health and stability
        - Provide BUY/HOLD/SELL recommendations with confidence scores
        
        Respond with structured analysis including numerical scores and clear reasoning.
        Focus on data-driven insights and quantitative evidence.
        """
    
    def analyze(self, company_symbol: str, market_data: Dict) -> AgentRecommendation:
        financial_metrics = market_data.get('financial_metrics', {}).get(company_symbol, {})
        
        prompt = f"""
        {self.get_system_prompt()}
        
        Analyze {company_symbol} based on these financial metrics:
        {financial_metrics}
        
        Provide:
        1. Recommendation (BUY/HOLD/SELL)
        2. Confidence level (HIGH/MEDIUM/LOW)
        3. Score (0-10)
        4. Brief reasoning
        
        Format: RECOMMENDATION|CONFIDENCE|SCORE|REASONING
        """
        
        try:
            response = self.llm.invoke(prompt)
            content = response.content if hasattr(response, 'content') else str(response)
            
            # Parse response (simplified)
            parts = content.split('|')
            if len(parts) >= 4:
                recommendation = parts[0].strip()
                confidence = parts[1].strip()
                score = float(parts[2].strip()) if parts[2].strip().replace('.', '').isdigit() else 5.0
                reasoning = parts[3].strip()
            else:
                recommendation = "HOLD"
                confidence = "MEDIUM"
                score = 5.0
                reasoning = "Analysis completed with default values"
            
            return AgentRecommendation(
                agent_role=self.role,
                company_symbol=company_symbol,
                recommendation=recommendation,
                confidence=DecisionConfidence(confidence.lower()),
                score=score,
                reasoning=reasoning,
                supporting_data=financial_metrics,
                timestamp=datetime.now().isoformat()
            )
        except Exception as e:
            # Fallback recommendation
            return AgentRecommendation(
                agent_role=self.role,
                company_symbol=company_symbol,
                recommendation="HOLD",
                confidence=DecisionConfidence.MEDIUM,
                score=5.0,
                reasoning=f"Error in analysis: {str(e)}",
                supporting_data={},
                timestamp=datetime.now().isoformat()
            )

print("✅ Agentic Decision System framework initialized!")
print("🤖 Base classes and Financial Analyst Agent defined")
print("=" * 60)
````

````python
# 🎯 Specialized Investment Agents Implementation
print("🎯 Implementing Specialized Investment Agents...")
print("=" * 60)

class RiskAssessmentAgent(AutonomousAgent):
    """Specialized agent for risk analysis"""
    
    def get_system_prompt(self) -> str:
        return """
        You are an expert Risk Assessment Agent specializing in investment risk analysis.
        
        Your responsibilities:
        - Evaluate volatility and downside risk
        - Assess maximum drawdown scenarios
        - Analyze risk-adjusted returns
        - Provide risk-aware investment recommendations
        
        Focus on risk mitigation and capital preservation.
        """
    
    def analyze(self, company_symbol: str, market_data: Dict) -> AgentRecommendation:
        financial_metrics = market_data.get('financial_metrics', {}).get(company_symbol, {})
        volatility = financial_metrics.get('volatility', 0)
        max_drawdown = financial_metrics.get('max_drawdown', 0)
        sharpe_ratio = financial_metrics.get('sharpe_ratio', 0)
        
        # Risk-based scoring logic
        risk_score = 5.0
        if volatility < 0.2 and max_drawdown > -0.1 and sharpe_ratio > 1.0:
            risk_score = 8.0
            recommendation = "BUY"
            confidence = DecisionConfidence.HIGH
        elif volatility < 0.3 and max_drawdown > -0.2:
            risk_score = 6.5
            recommendation = "HOLD"
            confidence = DecisionConfidence.MEDIUM
        else:
            risk_score = 3.0
            recommendation = "SELL"
            confidence = DecisionConfidence.MEDIUM
        
        reasoning = f"Risk assessment: Volatility {volatility:.2%}, Max DD {max_drawdown:.2%}, Sharpe {sharpe_ratio:.2f}"
        
        return AgentRecommendation(
            agent_role=self.role,
            company_symbol=company_symbol,
            recommendation=recommendation,
            confidence=confidence,
            score=risk_score,
            reasoning=reasoning,
            supporting_data=financial_metrics,
            timestamp=datetime.now().isoformat()
        )

class AIStrategistAgent(AutonomousAgent):
    """Specialized agent for AI strategy evaluation"""
    
    def get_system_prompt(self) -> str:
        return """
        You are an expert AI Strategist Agent specializing in artificial intelligence initiatives evaluation.
        
        Your responsibilities:
        - Evaluate AI innovation scores and strategic positioning
        - Assess long-term AI competitive advantages
        - Analyze AI initiative sustainability
        - Provide AI-focused investment recommendations
        
        Focus on future potential and technological leadership.
        """
    
    def analyze(self, company_symbol: str, market_data: Dict) -> AgentRecommendation:
        ai_metrics = market_data.get('ai_metrics', {}).get(company_symbol, {})
        ai_score = ai_metrics.get('ai_score', 0)
        
        # AI-focused scoring logic
        if ai_score >= 0.8:
            ai_strategy_score = 9.0
            recommendation = "BUY"
            confidence = DecisionConfidence.HIGH
        elif ai_score >= 0.6:
            ai_strategy_score = 7.0
            recommendation = "BUY"
            confidence = DecisionConfidence.MEDIUM
        elif ai_score >= 0.4:
            ai_strategy_score = 5.5
            recommendation = "HOLD"
            confidence = DecisionConfidence.MEDIUM
        else:
            ai_strategy_score = 3.0
            recommendation = "HOLD"
            confidence = DecisionConfidence.LOW
        
        reasoning = f"AI Strategy analysis: AI score {ai_score:.2f}, Strong innovation pipeline"
        
        return AgentRecommendation(
            agent_role=self.role,
            company_symbol=company_symbol,
            recommendation=recommendation,
            confidence=confidence,
            score=ai_strategy_score,
            reasoning=reasoning,
            supporting_data=ai_metrics,
            timestamp=datetime.now().isoformat()
        )

class PortfolioManagerAgent(AutonomousAgent):
    """Specialized agent for portfolio optimization"""
    
    def get_system_prompt(self) -> str:
        return """
        You are an expert Portfolio Manager Agent specializing in portfolio optimization and allocation.
        
        Your responsibilities:
        - Optimize portfolio allocation across multiple assets
        - Balance risk and return objectives
        - Consider correlation and diversification
        - Provide allocation-focused recommendations
        
        Focus on portfolio-level optimization and diversification benefits.
        """
    
    def analyze(self, company_symbol: str, market_data: Dict) -> AgentRecommendation:
        financial_metrics = market_data.get('financial_metrics', {}).get(company_symbol, {})
        returns = financial_metrics.get('annualized_return', 0)
        volatility = financial_metrics.get('volatility', 0)
        
        # Portfolio optimization logic
        risk_adjusted_return = returns / max(volatility, 0.01) if volatility > 0 else returns
        
        if risk_adjusted_return > 2.0:
            portfolio_score = 8.5
            recommendation = "BUY"
            confidence = DecisionConfidence.HIGH
        elif risk_adjusted_return > 1.0:
            portfolio_score = 6.5
            recommendation = "HOLD"
            confidence = DecisionConfidence.MEDIUM
        else:
            portfolio_score = 4.0
            recommendation = "HOLD"
            confidence = DecisionConfidence.LOW
        
        reasoning = f"Portfolio analysis: Risk-adj return {risk_adjusted_return:.2f}, Suitable for diversified portfolio"
        
        return AgentRecommendation(
            agent_role=self.role,
            company_symbol=company_symbol,
            recommendation=recommendation,
            confidence=confidence,
            score=portfolio_score,
            reasoning=reasoning,
            supporting_data=financial_metrics,
            timestamp=datetime.now().isoformat()
        )

print("✅ All specialized agents implemented!")
print("🤖 Financial Analyst, Risk Assessor, AI Strategist, Portfolio Manager")
print("=" * 60)
````

````python
# 🎭 Decision Orchestrator - Multi-Agent Coordination Engine
print("🎭 Building Decision Orchestrator...")
print("=" * 60)

class DecisionOrchestrator:
    """
    Central coordination engine for multi-agent investment decisions
    Implements consensus algorithms and decision fusion logic
    """
    
    def __init__(self, llm_client):
        self.llm = llm_client
        self.agents = []
        self.decision_history = []
        
    def register_agent(self, agent: AutonomousAgent):
        """Register a specialized agent with the orchestrator"""
        self.agents.append(agent)
        print(f"  ✓ Registered {agent.role.value} agent")
    
    def analyze_company(self, company_symbol: str, market_data: Dict) -> ConsensusDecision:
        """Coordinate multi-agent analysis for a company"""
        print(f"\n🔍 Analyzing {company_symbol} with {len(self.agents)} agents...")
        
        # Collect recommendations from all agents
        agent_recommendations = []
        for agent in self.agents:
            try:
                recommendation = agent.analyze(company_symbol, market_data)
                agent_recommendations.append(recommendation)
                print(f"  {agent.role.value}: {recommendation.recommendation} ({recommendation.score:.1f}/10)")
            except Exception as e:
                print(f"  ⚠️ {agent.role.value} failed: {e}")
        
        # Apply consensus algorithm
        consensus = self._build_consensus(company_symbol, agent_recommendations)
        self.decision_history.append(consensus)
        
        return consensus
    
    def _build_consensus(self, company_symbol: str, recommendations: List[AgentRecommendation]) -> ConsensusDecision:
        """Apply consensus algorithm to agent recommendations"""
        if not recommendations:
            return self._default_decision(company_symbol)
        
        # Analyze recommendation patterns
        buy_votes = sum(1 for r in recommendations if r.recommendation == "BUY")
        hold_votes = sum(1 for r in recommendations if r.recommendation == "HOLD")
        sell_votes = sum(1 for r in recommendations if r.recommendation == "SELL")
        
        # Calculate weighted consensus score
        total_score = sum(r.score for r in recommendations)
        avg_score = total_score / len(recommendations)
        
        # Determine final recommendation based on majority vote
        if buy_votes > hold_votes and buy_votes > sell_votes:
            final_recommendation = "BUY"
            allocation = min(0.25, avg_score / 40)  # Max 25% allocation
        elif sell_votes > hold_votes and sell_votes > buy_votes:
            final_recommendation = "SELL"
            allocation = 0.0
        else:
            final_recommendation = "HOLD"
            allocation = min(0.15, avg_score / 50)  # Conservative allocation
        
        # Assess risk level
        risk_scores = [r.score for r in recommendations if r.agent_role == AgentRole.RISK_ASSESSOR]
        risk_level = "MEDIUM"
        if risk_scores:
            if risk_scores[0] > 7:
                risk_level = "LOW"
            elif risk_scores[0] < 4:
                risk_level = "HIGH"
        
        # Generate consensus rationale
        rationale = self._generate_consensus_rationale(recommendations)
        
        return ConsensusDecision(
            company_symbol=company_symbol,
            final_recommendation=final_recommendation,
            consensus_score=avg_score,
            agent_agreements=max(buy_votes, hold_votes, sell_votes),
            total_agents=len(recommendations),
            risk_level=risk_level,
            allocation_percentage=allocation * 100,
            decision_rationale=rationale
        )
    
    def _generate_consensus_rationale(self, recommendations: List[AgentRecommendation]) -> str:
        """Generate human-readable consensus rationale"""
        agent_insights = []
        for rec in recommendations:
            agent_insights.append(f"{rec.agent_role.value}: {rec.reasoning}")
        
        return "; ".join(agent_insights)
    
    def _default_decision(self, company_symbol: str) -> ConsensusDecision:
        """Fallback decision when no agent recommendations available"""
        return ConsensusDecision(
            company_symbol=company_symbol,
            final_recommendation="HOLD",
            consensus_score=5.0,
            agent_agreements=0,
            total_agents=0,
            risk_level="UNKNOWN",
            allocation_percentage=0.0,
            decision_rationale="No agent recommendations available"
        )
    
    def generate_portfolio_report(self) -> str:
        """Generate comprehensive portfolio analysis report"""
        if not self.decision_history:
            return "No decisions available for portfolio report"
        
        report = "🎯 AUTONOMOUS PORTFOLIO RECOMMENDATIONS\n"
        report += "=" * 50 + "\n\n"
        
        total_allocation = 0
        for decision in self.decision_history:
            total_allocation += decision.allocation_percentage
            
            report += f"📊 {decision.company_symbol}:\n"
            report += f"   Recommendation: {decision.final_recommendation}\n"
            report += f"   Consensus Score: {decision.consensus_score:.1f}/10\n"
            report += f"   Agent Agreement: {decision.agent_agreements}/{decision.total_agents}\n"
            report += f"   Risk Level: {decision.risk_level}\n"
            report += f"   Allocation: {decision.allocation_percentage:.1f}%\n"
            report += f"   Rationale: {decision.decision_rationale[:100]}...\n\n"
        
        report += f"💼 Total Portfolio Allocation: {total_allocation:.1f}%\n"
        report += f"💰 Cash Reserve: {100 - total_allocation:.1f}%\n"
        
        return report

# Initialize the Decision Orchestrator
orchestrator = DecisionOrchestrator(llm)

# Register all specialized agents
financial_agent = FinancialAnalystAgent(AgentRole.FINANCIAL_ANALYST, llm_mini)
risk_agent = RiskAssessmentAgent(AgentRole.RISK_ASSESSOR, llm_mini)
ai_agent = AIStrategistAgent(AgentRole.AI_STRATEGIST, llm_mini)
portfolio_agent = PortfolioManagerAgent(AgentRole.PORTFOLIO_MANAGER, llm_mini)

orchestrator.register_agent(financial_agent)
orchestrator.register_agent(risk_agent)
orchestrator.register_agent(ai_agent)
orchestrator.register_agent(portfolio_agent)

print("\n✅ Decision Orchestrator initialized with 4 specialized agents!")
print("🎭 Ready for autonomous multi-agent investment analysis")
print("=" * 60)
````

```python
import pickle
from datetime import datetime, timedelta

class PersistentDualLensKG(DualLensKnowledgeGraph):
    def __init__(self, cache_file='duallens_kg.pkl', cache_ttl_hours=24):
        self.cache_file = cache_file
        self.cache_ttl = timedelta(hours=cache_ttl_hours)
        
        # Try loading from cache
        if self._is_cache_valid():
            self._load_from_cache()
        else:
            super().__init__()
            
    def _is_cache_valid(self):
        if not os.path.exists(self.cache_file):
            return False
        cache_time = datetime.fromtimestamp(os.path.getmtime(self.cache_file))
        return datetime.now() - cache_time < self.cache_ttl
    
    def _load_from_cache(self):
        with open(self.cache_file, 'rb') as f:
            data = pickle.load(f)
            self.graph = data['graph']
            self.node_data = data['node_data']
            
    def save_to_cache(self):
        with open(self.cache_file, 'wb') as f:
            pickle.dump({'graph': self.graph, 'node_data': self.node_data}, f)
```

```python
from dataclasses import dataclass
from typing import List, Dict

@dataclass
class AgentDecision:
    agent_name: str
    recommendation: str
    confidence: float
    reasoning: str
    supporting_data: Dict

class ExplainableMultiAgentSystem:
    def make_decision(self, ticker: str) -> Dict:
        decisions = []
        
        # Topology Agent
        topology_score = self._analyze_topology(ticker)
        decisions.append(AgentDecision(
            agent_name="Graph Topology Agent",
            recommendation="BUY" if topology_score > 0.7 else "HOLD",
            confidence=topology_score,
            reasoning=f"Centrality score {topology_score:.2f} indicates strong market position",
            supporting_data={"degree": 15, "betweenness": 0.8}
        ))
        
        # RAG Agent
        rag_response = self.rag.query(f"Should I invest in {ticker}?")
        decisions.append(AgentDecision(
            agent_name="RAG Query Agent",
            recommendation=self._parse_rag_recommendation(rag_response),
            confidence=0.85,
            reasoning=rag_response[:200],
            supporting_data={"full_response": rag_response}
        ))
        
        # Consensus calculation
        final_decision = self._calculate_consensus(decisions)
        
        return {
            "final_recommendation": final_decision,
            "agent_breakdown": [asdict(d) for d in decisions],
            "explanation": self._generate_explanation(decisions, final_decision)
        }
    
    def _generate_explanation(self, decisions: List[AgentDecision], final: str) -> str:
        explanation = f"**Final Decision: {final}**\n\n"
        explanation += "**Agent Analysis:**\n"
        for d in decisions:
            explanation += f"- **{d.agent_name}**: {d.recommendation} "
            explanation += f"(confidence: {d.confidence:.0%})\n"
            explanation += f"  - Reasoning: {d.reasoning}\n"
        return explanation
```

```python
        self.monitor_thread = threading.Thread(target=self._monitor_loop)
        self.monitor_thread.start()
        
    def _monitor_loop(self):
        while self.running:
            for node in self.kg.graph.nodes():
                current_risk = self._calculate_current_risk(node)
                historical_risk = self.kg.node_data[node].get('risk_score', 0.5)
                
                if current_risk - historical_risk > 0.2:  # 20% increase
                    self._trigger_alert(node, current_risk, historical_risk)
                    
            time.sleep(self.check_interval)
    
    def _trigger_alert(self, ticker, current_risk, previous_risk):
        alert = {
            'timestamp': datetime.now(),
            'ticker': ticker,
            'risk_level': self._categorize_risk(current_risk),
            'change': current_risk - previous_risk,
            'message': f"Risk escalation detected for {ticker}"
        }
        self.alerts.append(alert)
        self._send_notification(alert)  # Email/Slack/etc.
```

```python
import networkx as nx
from sklearn.ensemble import IsolationForest

class GraphAnalytics:
    def __init__(self, kg: DualLensKnowledgeGraph):
        self.kg = kg
        
    def detect_anomalies(self) -> List[str]:
        """Identify unusual nodes using graph metrics"""
        features = []
        nodes = []
        
        for node in self.kg.graph.nodes():
            degree = self.kg.graph.degree[node]  # type: ignore
            clustering = nx.clustering(self.kg.graph.to_undirected(), node)
            betweenness = nx.betweenness_centrality(self.kg.graph)[node]
            
            features.append([degree, clustering, betweenness])
            nodes.append(node)
        
        # Isolation Forest for anomaly detection
        clf = IsolationForest(contamination=0.1)
        predictions = clf.fit_predict(features)
        
        anomalies = [nodes[i] for i, pred in enumerate(predictions) if pred == -1]
        return anomalies
    
    def find_communities(self) -> Dict[int, List[str]]:
        """Detect market sectors/clusters"""
        undirected = self.kg.graph.to_undirected()
        communities = nx.community.louvain_communities(undirected)
        return {i: list(comm) for i, comm in enumerate(communities)}
    
    def predict_contagion_risk(self, failed_node: str, threshold=0.7) -> List[str]:
        """Simulate cascading failure from one node"""
        at_risk = []
        for node in self.kg.graph.nodes():
            if node == failed_node:
                continue
            # Check all paths from failed_node to node
            try:
                paths = list(nx.all_simple_paths(self.kg.graph, failed_node, node, cutoff=3))
                if len(paths) > 0:
                    risk_score = len(paths) / 10  # Normalize
                    if risk_score > threshold:
                        at_risk.append((node, risk_score))
            except nx.NetworkXNoPath:
                continue
        return sorted(at_risk, key=lambda x: x[1], reverse=True)
```
